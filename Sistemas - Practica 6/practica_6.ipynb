```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
colors = ["#006D2C", "#31AAAA","#74C476"]
df=pd.DataFrame({'year': ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018'], 
            'month' : ['Jan', 'Feb', 'Mar', 'Jan', 'Feb', 'Mar', 'Jan', 'Feb', 'Mar'], 
            'values': [5, 8, 12, 9, 7, 3, 1, 8, 10]})
colors = ["#006D2C", "#31AAAA", "#74C476"]
pivot_df = df.pivot(index = 'year', columns = 'month', values = 'values')
pivot_df.loc[:,['Jan', 'Feb', 'Mar']].plot.bar(stacked=True, color=colors, figsize=(10, 7))
```
